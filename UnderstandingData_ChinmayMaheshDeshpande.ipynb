#Exploratory Data Analysis

Exploratory Data Analysis (EDA) is a critical process in the data science and machine learning pipeline. It
involves summarizing the main characteristics of a dataset using both statistical and visualization techniques.
Introduced by John Tukey in the 1970s, EDA helps analysts understand data distributions, relationships be-
tween variables, and possible anomalies before applying predictive models.
EDA typically involves calculating summary statistics, such as mean, median, variance, and standard deviation,
along with graphical techniques like histograms, scatter plots, and box plots. These methods help in identifying
trends, patterns, and irregularities that may not be apparent in raw data. The process is iterative, requiring
continuous exploration and refinement to ensure data integrity before further analysis.

EDA performs several key functions that are essential in the data analysis workflow. First, it helps in un-
derstanding the overall structure of the dataset, ensuring that data is formatted correctly and free from in-
consistencies. Second, it assists in identifying missing values and outliers that may affect statistical modeling.
Detecting these irregularities early allows analysts to take corrective actions, such as imputing missing values
or transforming variables to improve model performance.

# Data Cleaning

Data cleaning is an essential step in data preprocessing to ensure the dataset is structured, consistent, and ready for analysis. This process helps remove inconsistencies, redundancies, and formatting issues that can affect the quality of insights drawn from the data.

### **Steps in Data Cleaning:**
1. **Removing Duplicates:**  
   - Duplicate rows can cause bias in analysis and inflate counts. We use `df.drop_duplicates(inplace=True)` to remove redundant entries.

2. **Standardizing Formats:**  
   - Text data often contains inconsistencies such as **extra spaces, mixed casing, or typos**.  
   - To maintain uniformity, categorical text columns (e.g., `title`, `director`, `rating`, etc.) are converted to **lowercase** and **trimmed of extra spaces**.

3. **Ensuring Correct Data Types:**  
   - Columns like `release_year` should be in **integer format**.  
   - The `date_added` column, originally a string, is converted into **datetime format** for easier analysis.  

Cleaning the data ensures a **well-structured, error-free dataset**, making further analysis, visualization, and model building more reliable.


```python
df_cleaned = df.copy()  # Create a cleaned version of the dataset

df_cleaned.drop_duplicates(inplace=True)

# 2. Standardizing text formats (e.g., trimming spaces and converting to lowercase for categorical columns)
text_columns = ['type', 'title', 'director', 'cast', 'country', 'rating', 'listed_in', 'description']
for col in text_columns:
    df_cleaned[col] = df_cleaned[col].str.strip().str.lower()

# 3. Ensuring correct data types
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)  # Already in integer format
df_cleaned['date_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce')  # Convert date_added again to ensure consistency

# Displaying cleaned dataset
# Display first few rows of the cleaned dataset
df_cleaned.head()
```

```python
missing_counts = df_cleaned.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]  # Filter only columns with missing values

plt.figure(figsize=(10, 6))
missing_counts.sort_values().plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Missing Values Per Column")
plt.xlabel("Count of Missing Values")
plt.ylabel("Columns")
plt.show()
```

# Handling Missing Values & Methods

Missing values in a dataset can arise due to various reasons, such as data entry errors, unavailability of information, or inconsistencies in data collection. Handling missing values is crucial to ensure the integrity and accuracy of data analysis. Different techniques can be applied based on the type of data and the extent of missing values:

### **Methods to Handle Missing Values**
1. **Mean Imputation**  Replacing missing values with the mean of the column is useful for **normally distributed** numerical data but can distort the dataset if outliers are present.
2. **Median Imputation**  More robust than the mean, replacing missing values with the **median** is effective when dealing with **skewed data** or outliers.
3. **Mode Imputation**  Used for **categorical data**, where missing values are replaced with the most frequently occurring value in the column.
4. **K-Nearest Neighbors (KNN) Imputation**  A more sophisticated approach that replaces missing values by finding the most similar instances in the dataset and averaging their values.
5. **Multiple Imputation by Chained Equations (MICE)**  This method models missing data based on other variables, making iterative predictions, and is often used in complex datasets.



```python
# Step 2: Handling Missing Values in df_cleaned with Imputation for 'date_added'

# Creating a copy to avoid modifying the original dataset
df_handled = df_cleaned.copy()

# 1. Filling missing categorical values with 'Unknown'
df_handled[['director', 'cast', 'country']] = df_handled[['director', 'cast', 'country']].fillna('Unknown')

# 2. Filling missing 'rating' with the most common value (mode)
if df_handled['rating'].isnull().sum() > 0:  # Check if there are missing values
    most_common_rating = df_handled['rating'].mode()[0]
    df_handled['rating'] = df_handled['rating'].fillna(most_common_rating)

# 3. Imputing missing 'date_added' with the most common date (mode)
if df_handled['date_added'].isnull().sum() > 0:  # Check if missing values exist
    most_common_date = df_handled['date_added'].mode()[0]
    df_handled['date_added'] = df_handled['date_added'].fillna(most_common_date)

df_handled['duration'].fillna('NA', inplace=True)

# Checking missing values after handling
missing_values_final = df_handled.isnull().sum()

# Display missing values summary
print("Final Missing Values After Handling:\n", missing_values_final)

# Optionally display the first few rows to confirm changes
df_handled.head()
```

# Handling Missing Values

Missing values can negatively impact data integrity, so appropriate techniques were applied to handle them systematically.

### ** Filling Missing Categorical Values**
- **Affected Columns:** `director`, `cast`, `country`
- **Action Taken:** Missing values in these categorical columns were replaced with **'Unknown'** to preserve all records while retaining meaningful information.

### ** Imputing Missing `rating`**
- **Affected Column:** `rating`
- **Action Taken:** Missing values in the `rating` column were filled with the **most common rating (mode)** to maintain consistency and avoid ambiguity.

### ** Handling Missing `date_added`**
- **Affected Column:** `date_added`
- **Action Taken:** The missing values were imputed with the **most frequently occurring date** to ensure logical coherence in content release timelines.

### ** Addressing Missing `duration`**
- **Affected Column:** `duration`
- **Action Taken:** Instead of removing records, missing values in the `duration` column were replaced with **'NA'** to indicate that the duration information was unavailable.

### **Final Outcome**
- After applying these strategies, **all missing values were handled**, ensuring a clean dataset for further analysis.

# **Outliers and Methods to Handle Them**

### **What Are Outliers?**
Outliers are data points that significantly **deviate from the expected range** of values in a dataset. These values may be unusually high or low compared to the rest of the data. Outliers can occur due to **natural variations**, **measurement errors**, or **data entry mistakes**. While they can sometimes indicate errors, they may also represent important insights.

### **How to Handle Outliers?**
There are several ways to handle outliers, depending on the dataset and the analysis goal:

1. **Detection Methods:**
   - **Boxplot Analysis:** Identifies values that lie beyond 1.5 times the interquartile range (IQR).
   - **Z-Score Method:** Flags data points that are more than a certain number of standard deviations away from the mean.
   - **Isolation Forest & DBSCAN:** Machine learning-based methods that detect anomalies in datasets.

2. **Handling Methods:**
   - **Removing Outliers:** If outliers are due to errors or extreme noise, they can be removed.
   - **Transforming Data:** Log transformation or scaling can be applied to reduce the effect of outliers.
   - **Capping Values:** Extreme values can be limited to a predefined threshold.
   - **Separating Outliers:** Instead of removing, a flag column can be created to filter them when necessary.

---

# **Bivariate & Multivariate Analysis in Exploratory Data Analysis (EDA)**

## **What is Bivariate & Multivariate Analysis?**

### **Bivariate Analysis**
Bivariate analysis involves examining **the relationship between two variables**. This helps in identifying **patterns, correlations, and dependencies** between different data attributes.  
Common techniques include:
- **Scatter plots**  To visualize relationships between two numerical variables.
- **Box plots**  To compare a numerical variable across different categories.
- **Crosstabs & bar plots**  To analyze relationships between two categorical variables.
- **Correlation heatmaps**  To measure the strength of numerical relationships.

### **Multivariate Analysis**
Multivariate analysis extends **beyond two variables** to **study multiple factors simultaneously**. This helps in understanding **complex relationships** in the data.  
Common techniques include:
- **Pair plots**  To visualize pairwise relationships across multiple numerical features.
- **Multivariate regression**  To predict outcomes based on multiple inputs.
- **Clustering & dimensionality reduction**  To group similar data points and reduce complexity.

---

## **Why is Bivariate & Multivariate Analysis Important in EDA?**
1. **Identifies Relationships:** Helps in understanding how variables interact, whether they are **positively or negatively correlated**.
2. **Feature Selection:** Detects **redundant or unimportant features**, improving model efficiency.
3. **Detects Patterns & Trends:** Reveals insights such as **seasonal trends, consumer behavior, or business impacts**.
4. **Enhances Data Understanding:** Aids in **data-driven decision-making**, improving model performance by selecting the most meaningful features.
5. **Supports Hypothesis Testing:** Allows analysts to validate **assumptions about relationships in the data** before model building.

---

# **Feature Engineering & Its Importance in Exploratory Data Analysis (EDA)**

### **What is Feature Engineering?**
Feature engineering is the process of **creating new features from existing data** to improve analysis and predictive modeling. It involves **transforming raw data into meaningful insights** that can enhance understanding and performance in data-driven decision-making.

### **Why is Feature Engineering Important in EDA?**
Feature engineering is a **crucial step in Exploratory Data Analysis (EDA)** because:
1. **Enhances Data Interpretability:** Extracting new variables like `content_age` helps in understanding patterns such as **how long after release Netflix acquires content**.
2. **Improves Data Quality:** Converting categorical variables (`listed_in`) into **one-hot encoded features** allows better filtering and analysis.
3. **Supports Trend Analysis:** Extracting `year_added` and `month_added` helps track **content acquisition trends**.
4. **Facilitates Better Filtering & Insights:** Creating **binary indicators (`is_old_film`, `is_recent`)** allows easy categorization of content into classic vs. modern.
5. **Prepares Data for Modeling:** Feature engineering lays the groundwork for **machine learning models** by providing structured, meaningful features.

### **Feature Engineering in Netflix Dataset**
In this dataset, feature engineering involved:
- Extracting **year and month** from `date_added` for content tracking.
- Calculating **`content_age`** to measure the time gap between release and Netflix addition.
- Creating **binary features** (`is_old_film`, `is_recent`) to categorize content based on recency.
- Extracting **numerical duration** from `duration` for structured analysis.
- **One-hot encoding genres** (`listed_in`) to allow filtering by genre.

By applying feature engineering, we enhance the datasets **analytical depth**, making it easier to **derive insights and perform predictive analysis**.


```python
import pandas as pd

# Step 6: Feature Engineering on df_handled

# Ensuring 'release_year' is numeric and has no missing values
df_handled['release_year'] = pd.to_numeric(df_handled['release_year'], errors='coerce')
df_handled = df_handled.dropna(subset=['release_year'])  # Remove rows where release_year is missing

# Ensuring 'date_added' is in datetime format
df_handled['date_added'] = pd.to_datetime(df_handled['date_added'], errors='coerce')

# 1. Extracting 'year_added' and 'month_added' from 'date_added'
df_handled['year_added'] = df_handled['date_added'].dt.year
df_handled['month_added'] = df_handled['date_added'].dt.month

# 2. Creating 'content_age' feature (how old the content was when added to Netflix)
df_handled['content_age'] = df_handled['year_added'] - df_handled['release_year']

# 3. Creating an indicator variable for old films (before 2000)
df_handled['is_old_film'] = df_handled['release_year'] < 2000

# 4. Extracting numerical duration (Movies: minutes, TV Shows: seasons)
df_handled['numeric_duration'] = df_handled['duration'].str.extract('(\d+)')
df_handled['numeric_duration'] = pd.to_numeric(df_handled['numeric_duration'], errors='coerce')

# 5. Creating a binary feature for recent content (released after 2015)
df_handled['is_recent'] = df_handled['release_year'] > 2015

# 6. One-Hot Encoding 'listed_in' (splitting genres into individual columns)
genres = df_handled['listed_in'].str.get_dummies(sep=', ')

# Ensuring proper data types before concatenation
df_handled = df_handled.convert_dtypes()
genres = genres.convert_dtypes()

# Concatenating one-hot encoded genres safely
df_handled = pd.concat([df_handled, genres], axis=1)

# Checking for any new invalid values or missing data after feature engineering
print("Missing values after feature engineering:\n", df_handled.isnull().sum())

# Display the first few rows to verify the new features
print("\nFeature-Engineered Dataset (First 5 Rows):")
print(df_handled.head())
```

# **Feature Engineering on the Netflix Dataset**

### **What is Feature Engineering?**
Feature engineering is the process of **creating new features from existing data** to improve analysis and predictive modeling. It transforms raw data into useful insights that enhance **data-driven decision-making**.

---

### **Steps We Took in Feature Engineering**

1. **Extracted `year_added` and `month_added` from `date_added`**  
   - We derived **year** and **month** from the `date_added` column to understand when the content was added to Netflix.

2. **Created `content_age` Feature**  
   - The **age of the content** was calculated by subtracting **`release_year`** from **`year_added`**.
   - This feature helps in understanding how long it takes for Netflix to acquire content after its release.

3. **Created an Indicator Variable `is_old_film`**  
   - This feature marks **content released before 2000** as **old films** (`True` if the release year is before 2000).
   - Helps in distinguishing between classic and modern content.

4. **Extracted Numerical Duration for Movies & TV Shows**  
   - We extracted the numeric portion from the **`duration`** column.
   - For **movies**, it represents the duration in minutes, and for **TV shows**, it represents the number of seasons.

5. **Created a Binary Feature `is_recent`**  
   - A new feature was created to flag **recent content** (released after 2015) as `True`, and older content as `False`.

6. **One-Hot Encoded the `listed_in` (Genres)**  
   - The `listed_in` column was split into individual genre columns.
   - This transformation allows us to analyze the presence of specific genres in the dataset.

---

### **Why is Feature Engineering Important in EDA?**
- **Improves Data Analysis:** By creating meaningful features, we can better understand **patterns**, **trends**, and **relationships** in the data.
- **Supports Better Decision Making:** Features like **`content_age`** and **`is_recent`** help to distinguish between classic and modern content, aiding Netflixs content strategy analysis.
- **Prepares Data for Predictive Modeling:** Well-engineered features help improve the **accuracy and interpretability** of machine learning models.
- **Enables More Detailed Insights:** Features such as **genre-based one-hot encoding** provide deeper insights into Netflix's content catalog.

---

### **Conclusion**
Feature engineering has transformed raw data into useful features like **content age**, **recent content flag**, and **one-hot encoded genres**, enhancing the datasets **analytical potential**. This is an essential step for both **understanding the data** and preparing it for **predictive analysis**.



# **Data Visualization & Its Importance in EDA**

### **What is Data Visualization?**
Data visualization is the graphical representation of data and its insights. It is a powerful tool used to understand complex datasets by representing key information in the form of charts, graphs, and plots. Visualizations help us to identify trends, patterns, outliers, and relationships between variables that might not be immediately obvious in raw data.

### **Why is Data Visualization Important in EDA?**
1. **Simplifies Complex Data**: Visualization helps to **simplify complex datasets** and transforms large volumes of raw data into a format that is easier to interpret.
2. **Identifies Patterns & Trends**: By visually exploring the data, we can quickly identify **patterns, trends**, and **correlations** between different variables.
3. **Highlights Outliers**: Visualizations such as boxplots or scatter plots help in detecting **outliers** and extreme values that may need further attention.
4. **Supports Data Cleaning**: It aids in identifying any **inconsistencies** or **errors** in the data, such as missing values or incorrect formats, which are crucial for data preparation.
5. **Facilitates Decision Making**: By presenting data in a **clear and intuitive manner**, it empowers data scientists and business stakeholders to make **data-driven decisions**.
